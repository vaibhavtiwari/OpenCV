```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('/content/group1.jfif')
# img = cv.resize(img,(img.shape[1]//4,img.shape[0]//4))
cv2_imshow(img)
```

```python
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
haar_cascade = cv.CascadeClassifier('/content/haar_face.xml')

faces_rect = haar_cascade.detectMultiScale(img_gray,scaleFactor=1.1,minNeighbors=1)

print(f'Number of faces found = {len(faces_rect)}')
```

```python
for (x,y,w,h) in faces_rect:
  cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),thickness=2)

cv2_imshow(img)
```
