# Resize and Rescale images and Videos

```python
 def rescaleFrame(frame, scale=0.75):
   #Images, videos and live images
   width = int(frame.shape[1]*scale)
   height = int(frame.shape[0]*scale)
   dimensions = (width,height)

   return cv.resize(frame,dimensions, interpolation=cv.INTER_AREA)
```

# Image Transformations

```python
#Translation
#number of pixels to be shift along the x axis and y axis
# -x --> Left
# -y --> Up
# x --> Right
# y --> Down
def translate(img,x,y):
  transMat = np.float32([[1,0,x],[0,1,y]])
  dimensions = (img.shape[1],img.shape[0]) # shape[0] --> height shape[1] --> width
  return cv.warpAffine(img,transMat,dimensions)
```

```python
img = cv.imread('/content/dog2.jfif')
img = rescaleFrame(img,scale=0.2)
img_translated = translate(img,-100,100)
cv2_imshow(img)
cv2_imshow(img_translated)
```

# Rotation

```python
img = cv.imread('/content/dog2.jfif')
img = rescaleFrame(img,scale=0.2)

def rotate(img,angle,rotPoint=None):
  (height,width) = img.shape[:2]

  if rotPoint is None:
    rotPoint = (width//2,height//2)

  rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1.0)
  dimensions = (width,height)

  return cv.warpAffine(img,rotMat,dimensions)

img_rotated = rotate(img,-60)
img_rotated_rotated = rotate(img_rotated,30)
cv2_imshow(img)
cv2_imshow(img_rotated)
cv2_imshow(img_rotated_rotated)
```
